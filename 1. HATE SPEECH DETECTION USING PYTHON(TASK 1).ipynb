```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
import re
import nltk
from nltk.util import re
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
stopword = set(stopwords.words("english"))
```

```python
df = pd.read_csv("/content/Hate Speech Detection using Python.csv")
df
```

```python
df['lables'] = df['class'].map({0:"Hate Speech Detected",1:"Offensive Language Detected",2:"No Hate",3:"No Hate and Offensive Speech"})
df
```

```python
df = df[['tweet','lables']]
df
```

```python
def clean(text):
  text=str(text).lower()
  text=re.sub('\[.*?\]','',text)
  text=re.sub('https?://\S+|www\.\S+','',text)
  text=re.sub('<.*?>+','',text)
  text=re.sub("[%s]" % re.escape(string.punctuation),'',text)
  text=re.sub('\n','',text)
  text=re.sub('\w*\d\w*','',text)
  text=[word for word in text.split(' ') if word not in stopword]
  text=" ".join(text)
  return text
df['tweet'] = df["tweet"].apply(clean)
df
```

```python
x=np.array(df["tweet"])
y=np.array(df["lables"])
cv=CountVectorizer()
x=cv.fit_transform(x)
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
clf=DecisionTreeClassifier()
clf.fit(X_train,y_train)
```

```python
test_data="I will kill you"
df=cv.transform([test_data]).toarray()
print(clf.predict(df))
```
